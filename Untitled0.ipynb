```python
import pandas as pd

df = pd.read_csv("messy_customer_data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#Data Cleaning
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
df['Gender'] = df['Gender'].str.strip().str.lower().map({
    'male': 'Male',
    'm': 'Male',
    'female': 'Female',
    'f': 'Female'
})
```

```python
df['Churn'] = df['Churn'].str.strip().str.lower().map({
    'yes': 'Yes',
    'y': 'Yes',
    'no': 'No',
    'n': 'No'
})
```

```python
df['TotalCharges'] = df['TotalCharges'].round(2)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['TotalCharges'] = df['TotalCharges'].round(2)
df['MonthlyCharges'] = df['MonthlyCharges'].round(2)
```

```python
# Tenure in years
df['Tenure_Years'] = df['Tenure'] / 12
```

```python
df['Tenure_Years'] = df['Tenure_Years'].round(2)
```

```python
# Churn flag
df['Churn_Flag'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
# Risk category
df['Risk'] = df.apply(lambda x: 'High Risk' if x['Tenure'] < 12 and x['MonthlyCharges'] > 70
                     else ('Medium Risk' if x['Tenure'] < 24 else 'Low Risk'), axis=1)
```

```python
##Basic Analysis
#churn%
(df['Churn'].value_counts(normalize=True) * 100).round(2)
```

```python
# Avg charges by churn
df.groupby('Churn')['MonthlyCharges'].mean().round(2)
```

```python
# Tenure vs churn
df.groupby('Churn')['Tenure'].mean().round(2)
```

```python
num_cols = df.select_dtypes(include=['float', 'int']).columns
df[num_cols] = df[num_cols].round(2)
```

```python
df.to_csv("final_cleaned_data.csv", index=False)
```
